```python
# import dependencies
import pandas as pd
import datetime as dt
```

```python
# attempt to increase col display
pd.set_option('display.max_columns', 28)
```

```python
# read in csv
policy_df = pd.read_csv('Raw/policy_list.csv')
policy_df.head()
```

```python
# viewing current columns
policy_df.columns
```

```python
# taking columns subset
policy_df = policy_df[['COUNTRY_NAME', 'ISO3', 'POLICY_TYPE', 'POLICY_SUBTYPE',
       'END_DATE', 'AIR', 'TARGETS_AIR', 'LAND',
        'TARGETS_LAND', 'SEA', 'TARGETS_SEA',
       'CITIZEN', 'CITIZEN_LIST', 'HISTORY_BAN', 'HISTORY_BAN_LIST', 'REFUGEE',
       'REFUGEE_LIST', 'VISA_BAN', 'VISA_BAN_LIST',
       'CITIZEN_EXCEP', 'CITIZEN_EXCEP_LIST', 'COUNTRY_EXCEP',
       'COUNTRY_EXCEP_LIST', 'WORK_EXCEP']]
policy_df.columns
```

```python
# renaming the columns
policy_df.columns = ['Country', 'Country_id', 'Policy_existing', 'Policy_info', 'End_date', 'Air_policy_existing',
                             'Air_policy_info', 'Land_policy_existing', 'Land_policy_info', 'Sea_policy_existing', 
                              'Sea_policy_info','Citizen_policy_existing', 'Citizen_policy_info', 
                              'History_policy_existing', 'History_policy_info', 'Refugee_policy_existing', 
                              'Refugee_policy_info', 'Visa_policy_existing', 'Visa_policy_info', 
                              'Citizen_policy_existing', 'Citizen_policy_info', 'Country_policy_existing', 
                              'Country_info', 'Work_policy_info']
policy_df.columns
```

```python
# re-viewing the columns
policy_df.head()
```

```python
# checking which cols have null values
policy_df.isnull().sum()
```

```python
# renaming NaNs
policy_df.fillna('no policy', inplace=True)
policy_df.head()
```

```python
# finding length of dataframe
len(policy_df)
```

```python
# finding number of countries in df
len(policy_df['Country_id'].unique())
```

```python
sorted_policy_df = policy_df.sort_values('Country_id')
sorted_policy_df.head(20)
```

```python
sorted_policy_df['End_date'].dtype
```

```python
dates_list = [x.replace('_', '-') for x in sorted_policy_df['End_date']]
dates_list_edited = ['01-01-01' if x == 'NONE' or x == '0' else x for x in dates_list]
dates_list_edited
```

```python
sorted_policy_df['End_date'] = dates_list_edited
sorted_policy_df.head()
```

```python
sorted_policy_df['End_date'] = pd.to_datetime(sorted_policy_df['End_date'])
```

```python
sorted_policy_df['End_date'].dtype
```

```python
double_sorted_policy_df = sorted_policy_df.sort_values(['Country_id','End_date'])
double_sorted_policy_df.head(20)
```

```python
last_date_df = double_sorted_policy_df.drop_duplicates(subset='Country_id', keep='last')
last_date_df.head(20)
```
