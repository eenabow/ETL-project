```python
# import dependencies
import pandas as pd
import datetime as dt
```

```python
# attempt to increase col display
pd.set_option('display.max_columns', 28)
```

```python
# read in csv
policy_df = pd.read_csv('Raw/policy_list.csv')
policy_df.sample(7)
```

```python
# viewing current columns
policy_df.columns
```

```python
# taking columns subset
policy_df = policy_df[['COUNTRY_NAME', 'ISO3', 'POLICY_TYPE', 'POLICY_SUBTYPE',
       'START_DATE', 'AIR', 'AIR_TYPE', 'TARGETS_AIR', 'LAND',
       'LAND_TYPE', 'TARGETS_LAND', 'SEA', 'SEA_TYPE', 'TARGETS_SEA',
       'CITIZEN', 'CITIZEN_LIST', 'HISTORY_BAN', 'HISTORY_BAN_LIST', 'REFUGEE',
       'REFUGEE_LIST', 'VISA_BAN', 'VISA_BAN_TYPE', 'VISA_BAN_LIST',
       'CITIZEN_EXCEP', 'CITIZEN_EXCEP_LIST', 'COUNTRY_EXCEP',
       'COUNTRY_EXCEP_LIST', 'WORK_EXCEP']]
policy_df.columns
```

```python
# renaming the columns
policy_df.columns = ['Country', 'Country_id', 'Policy_type', 'Policy_subtype', 'Policy_start_date', 'Air_ban',
                             'Air_ban_type', 'Air_ban_info', 'Land_ban', 'Land_ban_type', 'Land_ban_info', 
                             'Sea_ban', 'Sea_ban_type', 'Sea_ban_info','Citizen_ban', 'Citizen_ban_info', 
                             'Travel_history_ban', 'Travel_history_ban_info', 
                             'Refugee_ban', 'Refugee_ban_info', 
                             'Visa_ban', 'Visa_ban_list', 'Visa_ban_info', 
                             'Citizen_exceptions', 'Citizen_exceptions_info', 
                             'Country_exceptions', 'Country_exceptions_info', 'Work_exceptions']
policy_df.columns
```

```python
# re-viewing the columns
policy_df.head()
```

```python
# checking which cols have null values
policy_df.isnull().sum()
```

```python
# renaming NaNs
policy_df.fillna('no policy', inplace=True)
policy_df.head()
```

```python
# finding length of dataframe
len(policy_df)
```

```python
# finding number of countries in df
len(policy_df['Country_id'].unique())
```

```python
# sorting the country id alphabetically
sorted_policy_df = policy_df.sort_values('Country_id')
sorted_policy_df.head(20)
```

```python
# finding dtype for Start Date col
sorted_policy_df['Policy_start_date'].dtype
```

```python
# impossible date
sorted_policy_df[sorted_policy_df['Policy_start_date'] == '04_31_20']
```

```python
# formatting dates for use in datetime
# replacing nons and 0s with fake date 
dates_list = [x.replace('_', '-') for x in sorted_policy_df['Policy_start_date']]
dates_list_start = ['01-01-01' if x == 'NONE' or x == '0' else x for x in dates_list]
# educated guess as to end date of errant date shown in above cell
dates_list_edited = ['04-21-20' if x == '04-31-20' else x for x in dates_list_start]
```

```python
# replacing Start Date col w/ edited dates list
sorted_policy_df['Policy_start_date'] = dates_list_edited
sorted_policy_df.head()
```

```python
# converting Start Date col to dtype datetime
sorted_policy_df['Policy_start_date'] = pd.to_datetime(sorted_policy_df['Policy_start_date'])
```

```python
# checking dtype
sorted_policy_df['Policy_start_date'].dtype
```

```python
# sorting by country id then start date
double_sorted_policy_df = sorted_policy_df.sort_values(['Country_id','Policy_start_date'])
double_sorted_policy_df.head()
```

```python
# keeping only the rows with the most recent date for each country
last_date_df = double_sorted_policy_df.drop_duplicates(subset='Country_id', keep='last')
last_date_df.head()
```

```python
# len of df
len(last_date_df)
```

```python
# showing cols of df
last_date_df.columns
```

```python
# separating out only rows where a complete ban was in effect, saving only cols that apply
full_closures_df = last_date_df[last_date_df['Policy_type'] == 'COMPLETE'][['Country', 'Country_id', 'Policy_type', 'Policy_subtype', 
                                                                            'Policy_start_date', 'Citizen_exceptions', 'Citizen_exceptions_info', 
                                                                            'Country_exceptions',        'Country_exceptions_info', 'Work_exceptions']]
full_closures_df.reset_index(inplace=True, drop=True)
full_closures_df.head()
```

```python
# separating out only rows where a partial ban was in effect, saving only cols that apply
partial_closures_df = last_date_df[last_date_df['Policy_type'] == 'PARTIAL'][['Country', 'Country_id', 'Policy_type', 'Policy_subtype',
       'Policy_start_date', 'Air_ban', 'Air_ban_type', 'Air_ban_info',
       'Land_ban', 'Land_ban_type', 'Land_ban_info', 'Sea_ban', 'Sea_ban_type',
       'Sea_ban_info', 'Citizen_ban', 'Citizen_ban_info', 'Travel_history_ban',
       'Travel_history_ban_info', 'Refugee_ban', 'Refugee_ban_info',
       'Visa_ban', 'Visa_ban_list', 'Visa_ban_info']]
partial_closures_df.reset_index(inplace=True, drop=True)
partial_closures_df.head()
```

```python
# separating out only rows where no ban was in effect, saving only cols that apply
no_closures_df = last_date_df[last_date_df['Policy_type'] == 'NOPOLICYIMPLEMENTED'][['Country', 'Country_id', 'Policy_type', 'Policy_subtype']]
no_closures_df.reset_index(inplace=True, drop=True)
no_closures_df.head(10)
```

```python
# checking lengths of the 3 dfs
len(full_closures_df), len(partial_closures_df), len(no_closures_df)
```

```python
# creation of policy table
# lists of policy types, subtypes and ids
full_policy_subtypes_list = [x for x in full_closures_df['Policy_subtype'].unique()]
full_policy_types_list = ['Full' for x in full_closures_df['Policy_subtype'].unique()]
full_policy_ids_list = [f'F{x}' for x in range(1, len(full_policy_types_list)+1)]

partial_policy_subtypes_list = [x for x in partial_closures_df['Policy_subtype'].unique()]
partial_policy_types_list = ['Partial' for x in partial_closures_df['Policy_subtype'].unique()]
partial_policy_ids_list = [f'P{x}' for x in range(1, len(partial_policy_types_list)+1)]

no_policy_subtypes_list = [x for x in no_closures_df['Policy_subtype'].unique()]
no_policy_types_list = ['None' for x in no_closures_df['Policy_subtype'].unique()]
no_policy_ids_list = [f'N{x}' for x in range(1, len(no_policy_types_list)+1)]

all_policy_subtypes_list = full_policy_subtypes_list + partial_policy_subtypes_list + no_policy_subtypes_list
all_policy_types_list = full_policy_types_list + partial_policy_types_list + no_policy_types_list
policy_id_list = full_policy_ids_list + partial_policy_ids_list + no_policy_ids_list
all_policy_types_list, all_policy_subtypes_list, policy_id_list
```

```python
# creation of policy table
# creating df out of lists
policy_types_df = pd.DataFrame({
    'Policy_id': policy_id_list, 'Policy_type': all_policy_types_list, 'Policy_subtype': all_policy_subtypes_list})
policy_types_df
```

```python
# full closures -- country exceptions
full_country_df = pd.DataFrame(
    full_closures_df[full_closures_df['Country_exceptions'] == 1][['Country_id', 'Policy_start_date', 'Country_exceptions_info']]
)
full_country_df['Policy_id'] = 'F1'

# full closures -- work exceptions --! no further info on this closure type (no info col)
full_work_df = pd.DataFrame(
    full_closures_df[full_closures_df['Work_exceptions'] == 1][['Country_id', 'Policy_start_date']]
)
full_work_df['Policy_id'] = 'F2'

# full closures -- citizen exceptions
full_citizens_df = pd.DataFrame(
    full_closures_df[full_closures_df['Citizen_exceptions'] == 1][['Country_id', 'Policy_start_date', 'Citizen_exceptions_info']]
)
full_citizens_df['Policy_id'] = 'F3'

# full closures -- essential only --! no further info on this closure type (no info col)
full_essential_df = pd.DataFrame(
    full_closures_df[full_closures_df['Policy_subtype'] == 'ESSENTIAL_ONLY'][['Country_id', 'Policy_start_date']]
)
full_essential_df['Policy_id'] = 'F4'
```

```python
# partial closure -- border closure
# by this point refugee bans are no longer in the dataframe (bans have ended by that time point) 
# split further into air, land, sea
partial_borders_df = pd.DataFrame(
    partial_closures_df[partial_closures_df['Policy_subtype'] == 'BORDER_CLOSURE'][[
        'Country_id', 'Policy_start_date', 'Air_ban', 'Air_ban_type', 'Air_ban_info', 
        'Land_ban', 'Land_ban_type', 'Land_ban_info', 'Sea_ban', 'Sea_ban_type', 'Sea_ban_info']]
)
partial_borders_df['Policy_id'] = 'P1'
# # partial -- air
partial_air_df = pd.DataFrame(partial_borders_df[partial_borders_df['Air_ban'] == 1][[
    'Country_id', 'Policy_start_date', 'Air_ban', 'Air_ban_type', 'Air_ban_info', 'Policy_id']])
# # partial -- sea
partial_sea_df = pd.DataFrame(partial_borders_df[partial_borders_df['Sea_ban'] == 1][[
    'Country_id', 'Policy_start_date', 'Sea_ban', 'Sea_ban_type', 'Sea_ban_info', 'Policy_id']])
# # partial -- land
partial_land_df = pd.DataFrame(partial_borders_df[partial_borders_df['Land_ban'] == 1][[
    'Country_id', 'Policy_start_date', 'Land_ban', 'Land_ban_type', 'Land_ban_info', 'Policy_id']])

# partial closure -- visa ban
partial_visa_df = pd.DataFrame(
    partial_closures_df[partial_closures_df['Policy_subtype'] == 'VISA_BAN'][[
        'Country_id', 'Policy_start_date', 'Visa_ban', 'Visa_ban_list', 'Visa_ban_info']]
)
partial_visa_df['Policy_id'] = 'P2'

# partial closure -- citizenship ban
partial_citizenship_df = pd.DataFrame(
    partial_closures_df[partial_closures_df['Policy_subtype'] == 'CITIZENSHIP_BAN'][[
        'Country_id', 'Policy_start_date', 'Citizen_ban', 'Citizen_ban_info']]
)
partial_citizenship_df['Policy_id'] = 'P3'

# partial closure -- history ban
partial_history_df = pd.DataFrame(
    partial_closures_df[partial_closures_df['Policy_subtype'] == 'HISTORY_BAN'][[
        'Country_id', 'Policy_start_date', 'Travel_history_ban', 'Travel_history_ban_info']]
)
partial_history_df['Policy_id'] = 'P4'

partial_history_df.sample(5)
```

```python
# no closures, adding the policy type
no_closures_df['Policy_id'] = 'N1'
no_closures_df = no_closures_df[['Country_id', 'Policy_id']]
no_closures_df.head()
```

```python
# reading in csvs
vaccinations_df = pd.read_csv('Clean/vaccinations_cleaned.csv')
travel_df = pd.read_csv('Clean/travel_restrictions_clean.csv')
```

```python
# number of countries in each df - travel restrictions has the most
len(vaccinations_df['Country'].unique()), len(travel_df['Country'].unique()), len(last_date_df['Country'].unique())
```

```python
# creating a countries names and ids table
# subsetting each CSV source df for only country names and IDs
vacc_countries_df = vaccinations_df[['Country', 'Country_id']]
immig_countries_df = last_date_df[['Country', 'Country_id']]
travel_countries_df = travel_df[['Country', 'Country_id']]

# outer merging the travel restrictions and immigration restriction dfs 
# to capture country names not in the other table
first_countries_merge = travel_countries_df.merge(immig_countries_df, how='outer', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
print(len(first_countries_merge))

# outer merging the vaccinations and first merge dfs
# to capture country names not in the other table
full_merge = first_countries_merge.merge(vacc_countries_df, how='outer', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
print(len(full_merge))
full_merge.head()

```

```python
# checking to see if there are any duplicates in the merged tables -- there aren't
full_merge[full_merge.duplicated() == True]
```

```python
# making the table alphabetical, resetting the index, saving to a new descriptive name
countries_names_ids_df = full_merge.sort_values('Country').reset_index(drop=True)
countries_names_ids_df.head()
```

```python
# doing first merge w/ each of the 3 closure dfs, w/ the travel restrictions df
full_closures_merged = full_closures_df.merge(travel_df, how='inner', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
full_closures_merged.head()
```

```python
# doing first merge w/ each of the 3 closure dfs, w/ the travel restrictions df
partial_closures_merged = partial_closures_df.merge(travel_df, how='inner', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
partial_closures_merged.head()
```

```python
# doing first merge w/ each of the 3 closure dfs, w/ the travel restrictions df
no_closures_merged = no_closures_df.merge(travel_df, how='inner', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
no_closures_merged.head()
```

```python
# doing second merge w/ each of the 3 closure dfs, with vaccination data
full_closures_all = full_closures_df.merge(vaccinations_df, how='left', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
full_closures_all.sample(5)
```

```python
# doing second merge w/ each of the 3 closure dfs, with vaccination data
partial_closures_all = partial_closures_df.merge(vaccinations_df, how='left', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
partial_closures_all.sample(5)
```

```python
# doing second merge w/ each of the 3 closure dfs, with vaccination data
no_closures_all = no_closures_df.merge(vaccinations_df, how='left', on='Country_id', suffixes=('', '_DROP')).drop(columns='Country_DROP')
no_closures_all.head(10)
```

```python
# checking which cols have null values
# replacing nulls (only in vaccination sourced data cols) with 'no information'
# re-checking null values - none left
print(full_closures_all.isnull().sum(), partial_closures_all.isnull().sum(), no_closures_all.isnull().sum())
full_closures_all.fillna('no information', inplace=True)
partial_closures_all.fillna('no information', inplace=True)
no_closures_all.fillna('no information', inplace=True)
full_closures_all.isnull().sum(), partial_closures_all.isnull().sum(), no_closures_all.isnull().sum()
```

```python
# replacing the now 'no information' dates with placeholder date '01-01-01' in case datetimes need to be applied
# creating lists to hold the dates
# replacing the date cols with those lists
full_closures_dates = ['01-01-01' if x == 'no information' else x for x in full_closures_all['vaccination_date']]
partial_closures_dates = ['01-01-01' if x == 'no information' else x for x in partial_closures_all['vaccination_date']]
no_closures_dates = ['01-01-01' if x == 'no information' else x for x in no_closures_all['vaccination_date']]

full_closures_all['vaccination_date'] = full_closures_dates
partial_closures_all['vaccination_date'] = partial_closures_dates
no_closures_all['vaccination_date'] = no_closures_dates
```

```python
# exporting all policies df to csv
last_date_df.to_csv('Clean/policy_cleaned.csv', index=False)
```
